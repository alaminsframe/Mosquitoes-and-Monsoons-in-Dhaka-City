```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

#uploaded = files.upload()
```

```python
#Dengue data Cleaning
dengue = pd.read_csv('dengue_data_2025.csv')
final_dengue = dengue.drop(['Total'], axis=1)
final_dengue
```

```python
#Rainfall data Cleaning
rainfall = pd.read_csv('rainfall_data_2025.csv')
rainfall_data_null = rainfall[rainfall.isnull().any(axis=1)]
rainfall_data_null
```

```python
#Rainfall data Cleaning
manually_collected_rainfall = [0, 0, 0, 0, 0, 0, 0, 1, 8, 2, 0, 0, 0, 0, 0]
null_indices = [162, 163, 172, 173, 174, 175, 178, 179, 180, 186, 199, 200, 201, 222, 223]

rainfall.loc[null_indices, 'Rainfall_mm'] = manually_collected_rainfall
rainfall_data_null
rainfall.isnull().sum()
```

```python
#Weather data Cleaning
weather = pd.read_excel('weather_dhaka_2025_filtered.xlsx')

print(weather)
```

```python
#2025 data merged
final_dengue['Date'] = pd.to_datetime(final_dengue['Date'], format='%d/%m/%Y')
rainfall['Date'] = pd.to_datetime(rainfall['Date'], format='%d/%m/%Y')
weather['Date'] = pd.to_datetime(weather['Date'], format='%Y-%m-%d')

merged_df_2025 = final_dengue.merge(rainfall, on='Date').merge(weather, on='Date')

merged_df_2025 = round(merged_df_2025.sort_values(by='Date'))

print(merged_df_2025)
```

```python
#2020-2024 data
old_df = pd.read_csv('dhaka_dengue_weather_2021_2024.csv')
old_df['Date'] = pd.to_datetime(old_df['Date'], format='%d-%m-%y')  # dd-mm-yy format
old_df = round(old_df.sort_values(by='Date'))
old_df
```

```python
# finding missing values
full_date_range = pd.date_range(start='2020-01-01', end='2024-12-31')

# Find missing dates
missing_dates = full_date_range.difference(old_df['Date'])

# Output missing dates
print(f"Total missing dates: {len(missing_dates)}")
print(missing_dates)
```

```python
#adding missing dates


full_dates_df = pd.DataFrame({'Date': pd.date_range(start='2020-01-01', end='2024-12-31')})

# Merge your original data with the full date range
# This will introduce NaNs for any dates that were missing
merged_df = pd.merge(full_dates_df, old_df, on='Date', how='left')

# Sort by date again
merged_df = merged_df.sort_values(by='Date').reset_index(drop=True)

# View result
print(merged_df.isnull().sum())
```

```python
# adding data into null columns

# merged_df['District'].fillna('Dhaka', inplace=True)

# merged_df['Rainfall_mm'].fillna(0, inplace=True)

# merged_df['Dengue_Cases'].fillna(6, inplace=True)

merged_df['District'] = merged_df['District'].fillna('Dhaka')
merged_df['Rainfall_mm'] = merged_df['Rainfall_mm'].fillna(0)
merged_df['Dengue_Cases'] = merged_df['Dengue_Cases'].fillna(0)

#missing weathers for '2024-02-05', '2024-02-07', '2024-02-29', '2024-04-22' === 23/65, 21/69, 25/55, 33/72   Temperature/Humidity
merged_df.loc[merged_df['Date'] == '2024-02-05', 'Temperature'] = 23
merged_df.loc[merged_df['Date'] == '2024-02-05', 'Humidity'] = 65

merged_df.loc[merged_df['Date'] == '2024-02-07', 'Temperature'] = 21
merged_df.loc[merged_df['Date'] == '2024-02-07', 'Humidity'] = 69

merged_df.loc[merged_df['Date'] == '2024-02-29', 'Temperature'] = 25
merged_df.loc[merged_df['Date'] == '2024-02-29', 'Humidity'] = 55

merged_df.loc[merged_df['Date'] == '2024-04-22', 'Temperature'] = 33
merged_df.loc[merged_df['Date'] == '2024-04-22', 'Humidity'] = 72

print(merged_df)
```

```python
# Add missing columns to 2025 dataset

merged_df_2025['District'] = 'Dhaka'
merged_df_2025['Previous_Day_Dengue_Cases'] = merged_df_2025['Dengue_Cases'].shift(1)

# Reorder columns to match old_df
merged_df_2025 = merged_df_2025[['Date', 'District', 'Dengue_Cases', 'Temperature', 'Humidity', 'Rainfall_mm', 'Previous_Day_Dengue_Cases']]
merged_df_2025

```

```python
# Combine the cleaned 2020-2024 and 2025 datasets
full_df = pd.concat([merged_df, merged_df_2025], ignore_index=True)

# Sort by date to maintain order
full_df = full_df.sort_values(by='Date').reset_index(drop=True)
full_df
```

```python
# Checking for duplicates and nulls

full_df.isnull().sum()

full_df_with_null = full_df[full_df.isnull().any(axis=1)]

full_df.loc[full_df['Date'] == '2024-02-05', 'Previous_Day_Dengue_Cases'] = 9
full_df.loc[full_df['Date'] == '2024-02-07', 'Previous_Day_Dengue_Cases'] = 5
full_df.loc[full_df['Date'] == '2024-02-29', 'Previous_Day_Dengue_Cases'] = 3
full_df.loc[full_df['Date'] == '2024-04-22', 'Previous_Day_Dengue_Cases'] = 17
full_df.loc[full_df['Date'] == '2025-01-01', 'Previous_Day_Dengue_Cases'] = 30

full_df.isnull().sum()

full_df['Date'].duplicated().sum()
full_df = full_df.drop_duplicates(subset='Date', keep='last')
full_df
```

```python
# Download final data
round(full_df).to_csv('Dhaka_dengue_and_weather_2020_to_2025.csv', index=False)
files.download('Dhaka_dengue_and_weather_2020_to_2025.csv')
```
